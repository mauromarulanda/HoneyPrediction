```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.read_csv("./honeyproduction.csv")
data
```

```python
import matplotlib.pyplot as plt

plt.ylabel("honey production")
plt.xlabel("number colonies")
plt.scatter(data["numcol"],data["totalprod"],color="blue")
plt.show()
```

```python
from sklearn import linear_model

regresion = linear_model.LinearRegression()

colonies = data["numcol"].values.reshape((-1, 1))

modelo = regresion.fit(colonies, data["totalprod"])

print("IntersecciÃ³n (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)

entrada = [[0], [100000], [200000], [300000],[400000],[500000]]
modelo.predict(entrada)

plt.scatter(entrada, modelo.predict(entrada), color="red")
plt.plot(entrada, modelo.predict(entrada), color="red", linewidth = 3)

plt.ylabel("honey production (tons per month)")
plt.xlabel("number colonies")
plt.scatter(data["numcol"], data["totalprod"], color="blue", alpha=0.55)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.ylabel("honey production")
plt.xlabel("number colonies")
data["numcol"].plot(kind="bar", color="red", figsize=(10, 6))
data["totalprod"].plot(kind="line", color="blue", secondary_y=True)
plt.legend(["number colonies", "honey production"])
plt.title("Honey production vs number of colonies")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear un eje polar
plt.axes(projection="polar")

# Calcular el Ã¡ngulo y la distancia de cada punto
angle = np.linspace(0, 2*np.pi, len(data))
distance = data["totalprod"] / data["numcol"]

# Crear un grÃ¡fico de dispersiÃ³n en el eje polar
plt.scatter(angle, distance, color="green")

# Agregar etiquetas y tÃ­tulo
plt.xlabel("angle")
plt.ylabel("distance")
plt.title("Spiral plot of honey production vs number of colonies")
plt.show()
```
